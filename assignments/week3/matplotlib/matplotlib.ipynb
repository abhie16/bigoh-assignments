```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Line Plot

```python
x = np.arange(1,100)
y = x**2

plt.plot(x,y)

plt.xlabel('X axis')
plt.ylabel('y axis')
plt.title('simple line plot')
```

## SCatter plot

```python
x = np.random.rand(10)
y = np.random.rand(10)

plt.scatter(x,y,c='r')
```

## bar chart of sale data

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]

# Create a bar plot
plt.bar(categories, values, color='red')
```

## Histogram

```python
df1 = pd.read_csv('supermarket_sales - Sheet1.csv')
values = df1['gross income']

# what is bins 
# Create a histogram
plt.hist(values, color='green')
```

## Pie chart

```python
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 30, 20, 25]
explode = (0.1, 0, 0, 0)

# Create a pie chart
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%')


# Add a title
plt.title('Pie Chart')
```

## box plot

```python
data = np.random.rand(100)

plt.boxplot(data)
```

## stacked bar chart

```python
categories = ['A', 'B', 'C', 'D']
values1 = [3, 7, 2, 5]
values2 = [4, 6, 3, 4]

plt.bar(categories, values1,label='cat 1', color='b')
plt.bar(categories, values2,bottom=values1, label='cat 2', color='r')
```

## subplots

```python
data1 = np.random.randn(100)
data2 = np.random.randn(100)

# Create subplots
plt.subplot(2, 1,2)
plt.hist(data1, color='blue', edgecolor='black')
plt.title('Subplot 1')

plt.subplot(2, 1,1)
plt.scatter(data2, data1, color='red')
plt.title('Subplot 2')

# Adjust layout for better spacing
plt.tight_layout()
```

## Annotating Plots

```python
plt.annotate('Annotate here', xy=(0.3, 0.3), xytext=(0.4, 0.4),
             arrowprops=dict(facecolor='black', shrink=0.05))
```

## Error bar plot

```python
data = [1, 2, 3, 4]
errors = [0.1, 0.2, 0.15, 0.31]

# Create a plot with error bars
plt.errorbar(range(len(data)), data, yerr=errors, fmt='*', color='r', capsize=3)

# Add labels and title
plt.xlabel('Data Points')
plt.ylabel('Values')
plt.title('Error Bar Plot')
```

## 3D plot

```python
ax = plt.figure().add_subplot(projection='3d')

# Plot a sin curve using the x and y axes.
x = np.linspace(0, 1, 100)
print(x)
y = np.sin(x * 2 * np.pi) / 2 + 0.5
ax.plot(x, y, zs=0, zdir='z', label='curve in (x, y)')

# Plot scatterplot data (20 2D points per colour) on the x and z axes.
# colors = ('r', 'g', 'b', 'k')

# # Fixing random state for reproducibility
# np.random.seed(19680801)

# x = np.random.sample(20 * len(colors))
# y = np.random.sample(20 * len(colors))
# c_list = []
# for c in colors:
#     c_list.extend([c] * 20)
# # By using zdir='y', the y value of these points is fixed to the zs value 0
# # and the (x, y) points are plotted on the x and z axes.
# ax.scatter(x, y, zs=0, zdir='y', c=c_list, label='points in (x, z)')

# Make legend, set axes limits and labels
ax.legend()
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Customize the view angle so it's easier to see that the scatter points lie
# on the plane y=0
ax.view_init(elev=20., azim=-35)
```
