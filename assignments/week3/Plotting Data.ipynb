```python
import numpy as np
import matplotlib.pyplot as plt
import math as mth
```

```python
# Plot Sin
t =  np.arange(0,0.98 ,0.01)
x1 = 2 * mth.pi * 4 * t
y1 = np.sin(x1)
plt.plot(t,y1)
plt.show()
```

```python
# Plot Cos
y2 = np.cos(x1)
plt.plot(t, y2)
plt.show()
```

```python
# Combine sin and cos in one plot
plt.plot(t, y1)
plt.plot(t,y2)
plt.title('Sin Cos Variation Plot')
plt.xlabel('time')
plt.ylabel('value')
plt.legend(['sin(x)', 'cos(x)']) 
plt.show()
```

```python
# Subplot
plt.subplot(1,2,1) # 1 row, 2 columns, and this plot is the first plot
plt.plot(t,y1)
plt.title('sin')
plt.legend('sin')
plt.subplot(1,2,2) # 1 row, 2 columns, and this plot is the second plot.
plt.plot(t,y2)
plt.title('cos')
plt.legend('cos')
plt.show()
```
